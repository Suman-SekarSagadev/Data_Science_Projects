# Module 6: Assignment 1 - Data Visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import the mtcars.csv file
file_path = 'cars.csv'
mtcars_df = pd.read_csv(file_path)

# Generate a line plot graph for 'model' vs 'hp'
plt.figure(figsize=(12, 6))
plt.plot(mtcars_df['model'], mtcars_df['hp'], marker='o', linestyle='-', color='b', label='Horsepower')
plt.xlabel('Models of the Cars')
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs HorsePower')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```

# Module 6: Assignment 2 - Data Visualization 

```python
plt.figure(figsize=(10, 6))
plt.bar(mtcars_df['carb'], mtcars_df['gear'], color='orange', alpha=0.7)
plt.xlabel('Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('Carbs vs Gear')
plt.tight_layout()

# Display the plot
plt.show()
```

# Module 6: Assignment 3 - Data Visualization 

```python
plt.figure(figsize=(10, 6))
plt.hist(mtcars_df['wt'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight values')
plt.tight_layout()

# Display the plot
plt.show()
```

# Module 6: Assignment 4 - Data Visualization 

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
cyl_counts = mtcars_df['cyl'].value_counts()
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of cylinders')

# Plot a pie chart for 'model'
plt.subplot(1, 2, 2)
model_counts = mtcars_df['model'].value_counts()
plt.pie(model_counts, labels=model_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of car models')

# Display the plots
plt.tight_layout()
plt.show()
```

# Module 6: Assignment 5 - Data Visualization 

```python
plt.figure(figsize=(10, 6))
plt.fill_between(mtcars_df['am'], mtcars_df['carb'], color='skyblue', alpha=0.4)
plt.xlabel('Transmission (am)')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of Carburetors (Area Chart)')
plt.show()
```

# Data Visualization with Matplotlib Assignment 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cars data as a dataframe
cars_df = pd.read_csv('cars.csv')


cylinders_counts = cars_df['cyl'].value_counts()
cylinders_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Number of Cylinders')
plt.ylabel('Frequency of Cars')
plt.title('Number of Cylinders vs Frequency of Cars')
plt.show()

# Task 2: Bar graph of count of columns with null values
null_counts = cars_df.isnull().sum()
null_counts.plot(kind='bar', color='orange')
plt.xlabel('Columns')
plt.ylabel('Count of Null Values')
plt.title('Count of Null Values in Each Column')
plt.show()

# Task 3: Histogram for 'mpg' column
plt.hist(cars_df['mpg'], bins=20, color='green', edgecolor='black')
plt.xlabel('Miles per Gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.show()

# Task 4: Boxplot for 'hp' column
plt.boxplot(cars_df['hp'], vert=False)
plt.xlabel('Car Horsepower')
plt.title('Boxplot for Car Horsepower')
plt.show()
```
